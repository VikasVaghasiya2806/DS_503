```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from skimage import data
from skimage import img_as_ubyte

# Load the built-in 'cameraman' grayscale image (shape: 512x512)
gray_image = data.camera()  # Already grayscale

# Convert to uint8 (just to be safe for OpenCV-like processing)
gray_image = img_as_ubyte(gray_image)

# Apply PCA on rows: each row is a sample, each column is a feature
pca = PCA(n_components=50)  # Compress to 50 features
pca_data = pca.fit_transform(gray_image)

# Reconstruct image from PCA-reduced data
img_reconstructed = pca.inverse_transform(pca_data)

# Convert back to uint8 for image display
img_compress = np.clip(img_reconstructed, 0, 255).astype(np.uint8)

# Display original and compressed images
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.imshow(gray_image, cmap='gray')
plt.title("Original Cameraman")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(img_compress, cmap='gray')
plt.title("PCA Compressed (50 components)")
plt.axis("off")

plt.tight_layout()
plt.show()
```
